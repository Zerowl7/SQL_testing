```python
# Write your MySQL query statement below
SELECT id, 
(CASE
	WHEN p_id IS NULL THEN "Root"
    ELSE
        CASE
            WHEN id NOT IN (SELECT p_id FROM Tree WHERE p_id IS NOT NULL) THEN "Leaf"
            ELSE "Inner"
        END
 END) AS type
FROM Tree;
```
